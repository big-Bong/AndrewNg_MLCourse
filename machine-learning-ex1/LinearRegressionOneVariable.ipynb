```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data/ex1data1.txt",names=["population","profit"])
df
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(df['population'],df['profit'])
ax.set_xlabel('population')
ax.set_ylabel('profit')
plt.show()
```

```python
def computeCost(X,y,theta,M):
    h_theta = theta.dot(X)
    return (1/(2*M))*np.sum((h_theta - y)**2)
```

```python
def gradientDescent(X,y,theta,M,alpha=0.1):
    h_theta = theta.dot(X)
    theta = theta - (alpha/M)*(h_theta - y).dot(X.transpose())
    return theta
```

```python
def linearRegression(X,y,theta,M,iterations,alpha):
    for _ in range(iterations):
        #print(computeCost(X,y,theta,M))
        theta = gradientDescent(X,y,theta,M,alpha)
    return theta
```

```python
M = len(df)
df['X0'] = [1]*M
X = df[['X0','population']]
y = df[['profit']]
theta = np.array([[0.0,0.0]])
iterations = 1500
alpha = 0.01
theta = linearRegression(X.transpose().values,y.transpose().values,theta,M,iterations,alpha)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(df['population'],df['profit'])
ax.set_xlabel('population')
ax.set_ylabel('profit')
x_value=[x for x in range(25)]
y_value=[y*theta[0][1]+theta[0][0] for y in x_value]
plt.plot(x_value,y_value,color="r")
plt.show()
```
